__1.__ Plot the function $y=3x^2$ for $-1 \le x \le 3$ as a continuous line. Include enough points so that the curve you plot appears smooth. Label the axes $x$ and $y$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-1, 3, 100)
plt.plot(x, 3. * np.power(x, 2.), '-')
```

__3.__ Plot the functions $\sin x$ and $\cos x$ *vs* $x$ on the same plot with $x$ going from $-\pi$ to $\pi$. Make sure the limits of $x$-axis do not extend beyond the limits of the data. Plot $\sin x$ in the color green and $\cos x$ in the color black and include a legend to label the two curves. Place the legend within the plot, but such that it does not cover either of the sine or cosine traces.

```python
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x), 'g', label='sin')
plt.plot(x, np.cos(x), 'k', label='cos')
plt.legend()
```

__6.__ Plot force *vs* distance with error bars using the following data:

Your plot should also include a visual straight "best fit" to the data as well as visual "fits" that give the smallest and largest slopes consistent with the data. Note, you only need two points to define a straight line so the straight lines you draw on the plot should be arrays of length 2 and no longer. All of your fitted lines should lie *behind* the data. Try to make your plot look like the one below. *In addition*, add a legend to your plot the gives the slope with its uncertainty obtained from your visual fits to the data.

```python
d = np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18])
f = np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4])
df = np.array([ 0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4])
```

```python
plt.errorbar(d, f, fmt='o', yerr=df)
plt.ylabel('force'); plt.xlabel('distance')
d_fit = np.linspace(0.25, 2.25, 100)
a = 0.0
b = 3.3
f_fit = a + b * d_fit
plt.plot(d_fit, f_fit, 'r-', zorder=-1, label='best fit')
a = 0.0
b = 2.8
f_fit = a + b * d_fit
plt.plot(d_fit, f_fit, 'r:', zorder=-1, label='min', alpha=0.5)
a = 0.0
b = 3.8
f_fit = a + b * d_fit
plt.plot(d_fit, f_fit, 'r:', zorder=-1, label='max', alpha=0.5)
plt.legend();
```

```python
%cat ex7_data.txt
```

__(a)__ Use these data to calculate the velocity and acceleration. Be sure to calculate time arrays corresponding the midpoint in time between the two displacements or velocities for the velocity and acceleration arrays, respectively.

```python
time, pos, s_pos = np.loadtxt('ex7_data.txt', skiprows=4, unpack=True)
```

```python
v = (pos[1:] - pos[:-1]) / (time[1:] - time[:-1])
t_v = (time[1:] + time[:-1]) / 2.
```

```python
a = (v[1:] - v[:-1]) / (t_v[1:] - t_v[:-1])
t_a = (t_v[1:] + t_v[:-1]) / 2.
```

__(b)__ In a single window frame, make three vertically stacked plots of the displacement, velocity, and acceleration *vs* time. Show the error bars on the different plots. Make sure that the time axes of all three plots cover the same range of times. 

```python
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(6,10), sharex=True)

# subplot: linear plot of exponential
ax0.plot(time, pos, 'bo')
ax1.plot(t_v, v, 'bo')
ax2.plot(t_a, a, 'bo')
ax2.set_xlabel('Time (ms)')
ax0.set_ylabel('Distance (mm)')
ax1.set_ylabel('Velocity (mm/ms)')
ax2.set_ylabel('Acceleration (mm/ms$^2$)')
fig.tight_layout()
```
